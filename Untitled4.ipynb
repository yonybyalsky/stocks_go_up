```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def swap_rows(matrix, row, column):
    while row <= len(matrix) - 1:
        if matrix[row][column] == 0:
            another_row = row
            while another_row <= len(matrix) - 1:
                if matrix[another_row][column] != 0:
                    matrix[[row, another_row]] = matrix[[another_row, row]]
                another_row += 1
        row += 1
def gauess_method(matrix):
    row = 0
    while row <= len(matrix) - 1:
        swap_rows(matrix, row, row)
        matrix[row] *= (1 / matrix[row][row])
        another_row = 0
        while another_row <= len(matrix) - 1:
            if another_row != row and matrix[another_row][row] != 0:
                matrix[another_row] -= (matrix[another_row][row] * matrix[row]) 
            another_row += 1
        row += 1
    return matrix
```

```python
def generate_random_walk(n_steps: int, starting_point, drift, step_size) -> pd.DataFrame:
    value = starting_point
    time = 0
    data_arr = np.zeros((n_steps+1, 2))
    data_arr[0][0], data_arr[0][1] = time, value 
    for step in range(1, n_steps+1):
        time += step_size
        value += np.random.choice([-1, 1]) + drift
        data_arr[step][0], data_arr[step][1] = time, value 
    data = pd.DataFrame(data_arr, columns=['time', 'value'])
    return data
```

```python
data = generate_random_walk(500, 0, -0.001, 50)
with pd.ExcelWriter("path_to_file.xlsx") as writer:
    data.to_excel(writer)  
```

```python
plt.plot(data['time'], data['value'])
```

We can find the given parameters by finding the trendline of the graph (ax + b), "a" is the drift and "b" is the starting point. 

```python
def find_random_walk_parametes(random_walk_data):
    points_arr = random_walk_data.to_numpy()
    sum_x_points, sum_y_points = np.sum(points_arr, axis=0)
    n = len(points_arr)
    print(n)
    sum_x_mult_y, sum_x_power_2 = 0, 0
    for i in points_arr:
        sum_x_power_2 += i[0]**2 
        sum_x_mult_y += i[0]*i[1]
    arr = np.array([[sum_x_points, n, sum_y_points], [sum_x_power_2, sum_x_points, sum_x_mult_y]])
    arr = arr.astype('float64')
    arr1 = gauess_method(arr)
    return arr1[0, 2], arr1[1, 2]
```

```python
a, b = find_random_walk_parametes(data)
```
