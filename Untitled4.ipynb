```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas_datareader.data as web
import datetime
```

```python
def swap_rows(matrix, row, column):
    while row <= len(matrix) - 1:
        if matrix[row][column] == 0:
            another_row = row
            while another_row <= len(matrix) - 1:
                if matrix[another_row][column] != 0:
                    matrix[[row, another_row]] = matrix[[another_row, row]]
                another_row += 1
        row += 1
def gauess_method(matrix):
    row = 0
    while row <= len(matrix) - 1:
        swap_rows(matrix, row, row)
        matrix[row] *= (1 / matrix[row][row])
        another_row = 0
        while another_row <= len(matrix) - 1:
            if another_row != row and matrix[another_row][row] != 0:
                matrix[another_row] -= (matrix[another_row][row] * matrix[row]) 
            another_row += 1
        row += 1
    return matrix
```

```python
def generate_random_walk(n_steps: int, starting_point, drift, step_size) -> pd.DataFrame:
    value = starting_point
    time = 0
    value_per_time = np.zeros((n_steps+1, 2))
    value_per_time[0][0], value_per_time[0][1] = time, value 
    for step in range(1, n_steps+1):
        time += step_size
        value += np.random.choice([-1, 1]) + drift
        value_per_time[step][0], value_per_time[step][1] = time, value
    return pd.DataFrame(value_per_time, columns=['time', 'value'])
```

```python
data = generate_random_walk(2000, 21, -0.15, 23)
plt.plot(data['time'], data['value'])
```

We can find the given parameters by finding the trendline of the graph (ax + b), "a" is the drift and "b" is the starting point. 

```python
def find_trend_line_equation(data_frame):
    data_as_points = data_frame.to_numpy()
    sum_x_points, sum_y_points = np.sum(data_as_points, axis=0)
    sum_x_mult_y = np.sum(data_as_points[:,0]*data_as_points[:,1])
    sum_x_power_2 = np.sum(data_as_points[:,0]*data_as_points[:,0]) 
    matrix = np.array([[sum_x_points, len(data_as_points), sum_y_points], [sum_x_power_2, sum_x_points, sum_x_mult_y]]).astype('float64')
    solved_matrix = gauess_method(matrix)
    a = solved_matrix[0, 2]
    b = solved_matrix[1, 2]
    return a, b
```

```python
a, b = find_trend_line_equation(data)
plt.plot(data['time'], data['value'])
x = np.linspace(0, 2000*23, 2)
y = a*x + b  
plt.plot(x, y)
```
